```python

def get_imagenet_dataset(path_dir):
    
    if 'train' in path_dir: 
        raise ValueError('Specify the root directory not the train directory for the imagenet dataset')
    
    path_train = path_dir + '/train'
    path_val = path_dir + '/val'
    
    # specify image size ?????? -- lets set it to 224,224,3
    
    img_size = (224,224,3)
    
    train_dataset = tf.keras.utils.image_dataset_from_directory(path_train)
    val_dataset = tf.keras.utils.image_dataset_from_directory(path_val)
    
    return train_dataset,val_dataset
    
```

```python
train_dataset, test_dataset = get_imagenet_dataset(path_dir)
```
